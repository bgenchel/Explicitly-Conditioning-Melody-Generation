```python
%pylab inline

import json
import os
import os.path as op
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

matplotlib.rcParams['figure.figsize'] = [12, 10]

simple_dir = op.join(os.getcwd(), 'src', 'models', 'simple_pitch_duration', 'runs')
simple_pitch_train_losses = json.load(open(op.join(simple_dir, 'pitches', 'CP_First_Good_Run', 'train_losses.json'), 'r'))
simple_pitch_valid_losses = json.load(open(op.join(simple_dir, 'pitches', 'CP_First_Good_Run', 'valid_losses.json'), 'r'))
simple_dur_train_losses = json.load(open(op.join(simple_dir, 'durations', 'CP_First_Good_Run', 'train_losses.json'), 'r'))
simple_dur_valid_losses = json.load(open(op.join(simple_dir, 'durations', 'CP_First_Good_Run', 'valid_losses.json'), 'r'))

ccp_dir = op.join(os.getcwd(), 'src', 'models', 'chord_conditioned_pitch', 'runs')
ccp_pitch_train_losses = json.load(open(op.join(ccp_dir, 'pitches', 'CP_First_Good_Run', 'train_losses.json'), 'r'))
ccp_pitch_valid_losses = json.load(open(op.join(ccp_dir, 'pitches', 'CP_First_Good_Run', 'valid_losses.json'), 'r'))
ccp_dur_train_losses = json.load(open(op.join(ccp_dir, 'durations', 'CP_First_Good_Run', 'train_losses.json'), 'r'))
ccp_dur_valid_losses = json.load(open(op.join(ccp_dir, 'durations', 'CP_First_Good_Run', 'valid_losses.json'), 'r'))

ce_dir = op.join(os.getcwd(), 'src', 'models', 'condition_everything', 'runs')
ce_pitch_train_losses = json.load(open(op.join(ce_dir, 'pitches', 'CP_First_Good_Run', 'train_losses.json'), 'r'))
ce_pitch_valid_losses = json.load(open(op.join(ce_dir, 'pitches', 'CP_First_Good_Run', 'valid_losses.json'), 'r'))
ce_dur_train_losses = json.load(open(op.join(ce_dir, 'durations', 'CP_First_Good_Run', 'train_losses.json'), 'r'))
ce_dur_valid_losses = json.load(open(op.join(ce_dir, 'durations', 'CP_First_Good_Run', 'valid_losses.json'), 'r'))
```

```python
plt.title("No Conditioning - Pitch Net", fontsize=18)
plt.plot(simple_pitch_train_losses, 'b')
plt.plot(simple_pitch_valid_losses, 'r')
plt.xlabel('Epochs', fontsize=18)
plt.ylabel('NLL Loss', fontsize=18)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1, 1.2, 1.4, 1.6, 1.8, 2, 2.2, 2.4, 2.6, 2.8])

blue_patch = mpatches.Patch(color='blue', label='Training Loss')
red_patch = mpatches.Patch(color='red', label='Validation Loss')
plt.legend(handles=[blue_patch, red_patch])

plt.show()

plt.title("No Conditioning - Duration Net", fontsize=18)
plt.plot(simple_dur_train_losses, 'b')
plt.plot(simple_dur_valid_losses, 'r')
plt.xlabel('Epochs', fontsize=18)
plt.ylabel('NLL Loss', fontsize=18)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1, 1.2, 1.4, 1.6, 1.8, 2, 2.2, 2.4, 2.6, 2.8])


blue_patch = mpatches.Patch(color='blue', label='Training Loss')
red_patch = mpatches.Patch(color='red', label='Validation Loss')
plt.legend(handles=[blue_patch, red_patch])

plt.show()
```
